## Convert timestamps and align the datasets by date

Convert Timestamp IST to datetime

```python
historical_df['trade_datetime'] = pd.to_datetime(historical_df['Timestamp IST'], format='%d-%m-%Y %H:%M')
historical_df['date'] = historical_df['trade_datetime'].dt.date
```

```python
fear_greed_df['date'] = pd.to_datetime(
    fear_greed_df['date']
).dt.date
```

```python
merged_df = pd.merge(
    historical_df,
    fear_greed_df[['date', 'classification', 'value']],
    on='date',
    how='left'
)
```

## Create key metrics

Daily PnL per Trader (Account)

```python
daily_pnl = (
    merged_df
    .groupby(['Account', 'date'])['Closed PnL']
    .sum()
    .reset_index(name='daily_pnl')
)

daily_pnl.head()
```

Win Rate per Trader

```python
win_rate = (
    merged_df
    .assign(win=merged_df['Closed PnL'] > 0)
    .groupby('Account')['win']
    .mean()
    .mul(100)
    .reset_index(name='win_rate_pct')
)

win_rate.head()
```

Average Trade Size

```python
avg_trade_size = (
    merged_df
    .groupby('Account')['Size USD']
    .mean()
    .reset_index(name='avg_trade_size_usd')
)

avg_trade_size.head()
```

Number of Trades per Day

```python
trades_per_day = (
    merged_df
    .groupby('date')
    .size()
    .reset_index(name='num_trades')
)

trades_per_day.head()
```

Long / Short Ratio

```python
long_short_ratio = (
    merged_df['Direction']
    .value_counts(normalize=True)
    .mul(100)
)

print(long_short_ratio)
```

Leverage Distribution

```python
merged_df['estimated_leverage'] = (
    merged_df['Size USD'] /
    merged_df['Start Position'].replace(0, pd.NA)
)

merged_df['estimated_leverage'].describe()
```

These metrics provide a quantitative view of trader behavior and performance.

# Part B Anlysiss

Fear vs Greed Performance

Group sentiments into Fear vs Greed

```python
merged_df['sentiment_group'] = merged_df['classification'].replace({
    'Extreme Fear': 'Fear',
    'Fear': 'Fear',
    'Greed': 'Greed',
    'Extreme Greed': 'Greed'
})

merged_df['is_win'] = merged_df['Closed PnL'] > 0
```

PnL and Win Rate by Sentiment

```python
performance_by_sentiment = (
    merged_df
    .groupby('sentiment_group')
    .agg(
        avg_pnl=('Closed PnL', 'mean'),
        total_pnl=('Closed PnL', 'sum'),
        win_rate=('is_win', 'mean'),
        num_trades=('Closed PnL', 'count')
    )
)

performance_by_sentiment['win_rate'] *= 100

performance_by_sentiment
```

Use losing trades as a simple drawdown proxy.

```python
drawdown_proxy = (
    merged_df
    .groupby('sentiment_group')
    .agg(
        avg_loss=('Closed PnL',
                  lambda x: x[x < 0].mean())
    )
)

drawdown_proxy
```

Trading Behavior by Sentiment

Trade Frequency

```python
trade_frequency = (
    merged_df
    .groupby(['date','sentiment_group'])
    .size()
    .reset_index(name='trades')
)

trade_frequency.groupby('sentiment_group')['trades'].mean()
```

Position Size

```python
position_size = (
    merged_df
    .groupby('sentiment_group')['Size USD']
    .mean()
)

position_size
```

Long / Short Bias

```python
long_short = pd.crosstab(
    merged_df['sentiment_group'],
    merged_df['Direction'],
    normalize='index'
) * 100

long_short
```

Trader Segmentation

High vs Low Leverage

```python
merged_df['estimated_leverage'] = (
    merged_df['Size USD']
    /
    merged_df['Start Position'].replace(0, pd.NA)
)

median_leverage = merged_df['estimated_leverage'].median()

merged_df['leverage_segment'] = merged_df[
    'estimated_leverage'
].apply(
    lambda x: 'High Leverage'
    if pd.notna(x) and x > median_leverage
    else 'Low Leverage'
)
```

```python
merged_df.groupby('leverage_segment').agg(
    avg_pnl=('Closed PnL','mean'),
    win_rate=('is_win','mean')
)
```

Frequent vs Infrequent Traders

```python
trade_counts = merged_df.groupby('Account').size()

threshold = trade_counts.median()

merged_df['frequency_segment'] = merged_df['Account'].map(
    lambda x:
    'Frequent'
    if trade_counts[x] > threshold
    else 'Infrequent'
)
```

```python
merged_df.groupby('frequency_segment').agg(
    avg_pnl=('Closed PnL','mean'),
    win_rate=('is_win','mean')
)
```

Consistent vs Inconsistent Traders

```python
account_stats = merged_df.groupby('Account').agg(
    avg_pnl=('Closed PnL','mean'),
    pnl_std=('Closed PnL','std')
)

threshold_std = account_stats['pnl_std'].median()

account_stats['consistency'] = account_stats[
    'pnl_std'
].apply(
    lambda x:
    'Consistent'
    if x < threshold_std
    else 'Inconsistent'
)

account_stats.head()
```

# Part C — “Actionable output”

Adjust Leverage Based on Market Sentiment

The analysis shows that traders generally perform worse during Fear periods. Average profit (PnL) is lower, win rates decrease, and losses are larger compared to Greed periods. This means that using high leverage during Fear can increase risk and lead to bigger losses.

traders tend to increase their position sizes and trade more frequently during Fear periods, even though their performance is weaker. This suggests that traders may be taking on more risk when market conditions are less favorable.

**Rule of Thumb:**

During Fear or Extreme Fear

* Reduce leverage to limit potential losses.
* Trade less frequently and focus only on high-quality opportunities.
* Keep position sizes smaller to manage risk.

During Greed or Extreme Greed

* Use leverage more carefully when market conditions are favorable.
* Increase trading activity only if performance remains strong.
* Consider larger positions when win rates and profitability are higher.

By reducing risk during Fear periods and taking advantage of stronger performance during Greed periods, traders can improve capital preservation and potentially achieve better long-term results.


Increase Trading Activity Only When Market Sentiment is Favorable

The analysis shows that traders perform better during Greed periods. Average profits and win rates are higher when market sentiment is positive. This suggests that traders should be more active when the market environment is favorable.

However, the data also shows that traders currently do the opposite. During Fear periods, they trade much more frequently and use larger position sizes, even though their performance is weaker and losses are larger.

Another important finding is that less active traders tend to earn higher average profits than very active traders. This suggests that being selective and waiting for better opportunities can lead to better results.
